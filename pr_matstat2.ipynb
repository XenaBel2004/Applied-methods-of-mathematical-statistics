```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
n_values = [10**2, int(10**(2.5)), 10**3, int(10**(3.5)), 10**4, int(10**(4.5)), 10**5]

var_q01 = []
var_median = []
var_mean = []

for i in n_values:
    q_01 = []
    medians = []
    means = []
    for k in range(N):
        samples = np.random.uniform(0, 1, i)
        q_01.append(np.quantile(samples, 0.01))
        medians.append(np.median(samples))
        means.append(np.mean(samples))
    var_q01.append(np.var(q_01))
    var_median.append(np.var(medians))
    var_mean.append(np.var(means))

ln_n = np.log(n_values)

plt.figure(figsize=(8,6))
plt.plot(ln_n, var_q01, marker='o', label=" 0.01")
plt.plot(ln_n, var_median, marker='s', label="")
plt.plot(ln_n, var_mean, marker='^', label="")

plt.xlabel("ln(n)")
plt.ylabel(" ")
plt.title("    ln(n)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
m2_q01 = []
m2_median = []
m2_mean = []

for n in n_values:
    q_01 = []
    medians = []
    means = []
    for _ in range(N):
        samples = np.random.uniform(0, 1, n)
        q_01.append(np.quantile(samples, 0.01))
        medians.append(np.median(samples))
        means.append(np.mean(samples))

    m2_q01.append(np.mean(np.square(q_01)))
    m2_median.append(np.mean(np.square(medians)))
    m2_mean.append(np.mean(np.square(means)))

ln_n = np.log(n_values)

plt.figure(figsize=(8,6))
plt.plot(ln_n, m2_q01, 'o-', label=" 0.01")
plt.plot(ln_n, m2_median, 's-', label="")
plt.plot(ln_n, m2_mean, '^-', label="")

plt.xlabel("ln(n)")
plt.ylabel("  $")
plt.title("     ln(n)")
plt.legend()
plt.grid(True)
plt.show()
```
